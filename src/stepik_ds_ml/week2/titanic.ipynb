```python
import pandas as pd
from sklearn import tree
import numpy as np

```

```python
tit_data = pd.read_csv('train.csv')
tit_data.head()
tit_data.isnull().sum()
```

```python
X = tit_data.drop(['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'], axis=1)
y = tit_data['Survived']
#str -> int
X = pd.get_dummies(X)
#Nan -> median
X = X.fillna({'Age': X.Age.median()})


X.head()

```

```python
from sklearn.model_selection import train_test_split
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3)

X_train, x_test, y_train, y_test= train_test_split(X,y,test_size=0.33, random_state=42)


clf.fit(X_train, y_train)

clf.score(X_train, y_train)
```

```python
clf.score(x_test, y_test)

test_data = pd.read_csv('test.csv')
X = test_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
X = pd.get_dummies(X)
y = clf.predict(X)
df = pd.DataFrame()
df['PassengerId'] = test_data['PassengerId']
df['Survived'] = np.asarray(y).T

df.head()

df.to_csv('out.csv', index=False)
```

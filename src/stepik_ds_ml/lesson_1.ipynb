```python
wr_mean = student_perf['writing score'].mean()

student_perf.loc[(student_perf['writing score'] > wr_mean) & (student_perf['gender'] == 'female')]
```

```python
reduced_= student_perf.loc[(student_perf['lunch'] == 'free/reduced')]
mean_ms =  reduced_['math score'].mean()
mean_rs =  reduced_['reading score'].mean()
mean_ws =  reduced_['writing score'].mean()

standard = student_perf.loc[(student_perf['lunch'] == 'standard')]

print(standard['math score'].mean() > mean_ms)
print(standard['reading score'].mean() > mean_rs)
print(standard['writing score'].mean() > mean_ws)

print(standard['math score'].var() > reduced_['math score'].var())
print(standard['reading score'].var() > reduced_['reading score'].var())
print(standard['writing score'].var() > reduced_['writing score'].var())
```

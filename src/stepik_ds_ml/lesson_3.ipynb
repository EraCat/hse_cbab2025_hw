```python
from sklearn import tree
import pandas as pd
```

```python
data = pd.DataFrame(

{

"X_1": ["1", "1", "1", "0", "0", "0", "0", "1"],

"X_2": ["0", "0", "0", "1", "0", "0", "0", "1"],

"Y": ["1", "1", "1", "1", "0", "0", "0", "0"],

}

)
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
X = data[['X_1', 'X_2']]
y = data.Y
```

```python
clf.fit(X, y)

tree.plot_tree(clf)
```
